# 1. Hiển thị toàn bộ dữ liệu

```python
import pandas as pd
```

```python
import numpy as np
```

```python
data = pd.read_csv('P4AI_BT1.csv')
```

```python
data
```

# 2. Điền các giá trị bị thiếu của dữ liệu theo phương pháp  trung bình với các trường có giá trị số và theo phương pháp tần số cao nhất với giá trị rời rạc

```python
df1 = data.copy() # sao chép dữ liệu giúp thuận tiện cho việc sửa lỗi
```

```python
df1['sepal.length'].fillna(round(df1['sepal.length'].mean()), inplace=True)
df1['sepal.width'].fillna(round(df1['sepal.width'].mean()), inplace=True)
df1['petal.length'].fillna(round(df1['petal.length'].mean()), inplace=True)
df1['petal.width'].fillna(round(df1['petal.width'].mean()), inplace=True)
df1['variety'].fillna(df1['variety'].mode()[0], inplace = True)
```

# 3. Hiển thị các bản ghi có sepal.length > 5 và sepal.width > 3

```python
df1[(df1['sepal.length'] > 5) & (df1['sepal.width'] > 3)]
```

# 4. Chuẩn hóa dữ liệu theo min max

```python
df2 = df1.copy()
```

```python
sepal_length_min = df2['sepal.length'].min()
sepal_length_max = df2['sepal.length'].max()

sepal_width_min = df2['sepal.width'].min()
sepal_width_max = df2['sepal.width'].max()

petal_length_min = df2['petal.length'].min()
petal_length_max = df2['petal.length'].max()


petal_width_min = df2['petal.width'].min()
petal_width_max = df2['petal.width'].max()
```

```python
df2['sepal.length'] = (df2['sepal.length'] - sepal_length_min)/(sepal_length_max - sepal_length_min)
df2['sepal.width'] = (df2['sepal.width'] - sepal_width_min)/(sepal_width_max - sepal_width_min)
df2['petal.length'] = (df2['petal.length'] - petal_length_min)/(petal_length_max - petal_length_min)
df2['petal.width'] = (df2['petal.width'] - petal_width_min)/(petal_width_max - petal_width_min)
```

```python
df2
```

# 5. Liên tục hóa thuộc tính variety

```python
df3 = df2.copy()
```

```python
df3['variety'].unique()
```

```python
dummies = pd.get_dummies(df3['variety']).astype(int)
display(dummies)
```

```python
df3 = pd.concat([df3,dummies], axis=1)
# df3.drop('variety', axis=1, inplace=True)
df3
```

# 6. Rời rạc hóa thuộc tính sepal.length 

```python
df4 = df3.copy()
```

```python
df4['sepal_length'] = pd.cut(df4["sepal.length"],
       bins=[-0.1, 0.2, 0.4, 0.6, 0.8, 1],
       labels=["Very Low", "Low", "Medium", "High", "Very High"])
df4
```

# 7. Lấy mẫu ngẫu nhiên có lặp lại 50% dữ liệu


```python
sample = df4.sample(frac=0.5, replace=True, random_state=42)
len(sample)
sample
```

# 8. Xây dựng mô hình hồi quy tuyến tính để đoán kết quả của thuộc tính variety  dựa trên các thuộc tính khác

```python
df5 = df3.copy()
df5.variety.unique()  
```

```python
df5['variety'].replace({"Setosa":1,"Virginica":2, "Versicolor":3},inplace=True)
df5
```

```python
Train = df5.drop("variety", axis=1)
Test = df5["variety"]
display(Train)
display(Test)
```

```python
tp = int(0.9 * len(Train)) #thread point

X_train = Train[:tp]
Y_train = Test[:tp]
X_test = Train[tp:]
Y_test = Test[tp:]
```

```python
pip install scikit-learn
```

```python
from sklearn.linear_model import RidgeClassifier

reg = RidgeClassifier().fit(X_train, Y_train)
pred = reg.predict(X_test)
```

```python
reg.score(X_test,Y_test)
```

```python
target_class = pd.DataFrame({'Test values': pred,'True value': Y_test})
target_class
```

# 9. Sử dụng PCA và t-SNE để giảm chiều dữ liệu xuống 2 chiều và hiển thị dữ liệu trong không gian 2 chiều

```python
pip install matplotlib
```

```python
df6 = df5.copy()
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
# Định nghĩa DataFrame của bạn (giả sử là df)

# Áp dụng PCA để giảm chiều dữ liệu xuống 2 chiều
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df6)

# Áp dụng t-SNE để giảm chiều dữ liệu xuống 2 chiều
tsne = TSNE(n_components=2, random_state=42)
tsne_result = tsne.fit_transform(df6)

# Hiển thị kết quả PCA
plt.figure(figsize=(8, 6))
plt.scatter(pca_result[:, 0], pca_result[:, 1])
plt.title("PCA")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()

# Hiển thị kết quả t-SNE
plt.figure(figsize=(8, 6))
plt.scatter(tsne_result[:, 0], tsne_result[:, 1])
plt.title("t-SNE")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.show()
```

# 10. Phân cụm dữ liệu sử dụng DBSCAN trên dữ liệu 2 chiều và loại bỏ ngoại lai 


```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
df7 = df6.copy()
```

```python
# Định nghĩa DataFrame của bạn (giả sử là df)

# Chuẩn hoá dữ liệu
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df7)

# Áp dụng DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
cluster_labels = dbscan.fit_predict(scaled_data)

# Loại bỏ ngoại lai
df_filtered = df7[cluster_labels != -1]

# Hiển thị kết quả
plt.figure(figsize=(8, 6))
plt.scatter(df_filtered["sepal.length"], df_filtered["sepal.width"], c=cluster_labels[cluster_labels != -1])
plt.title("DBSCAN Clustering")
plt.xlabel("sepal.length")
plt.ylabel("sepal.width")
plt.show()
```
